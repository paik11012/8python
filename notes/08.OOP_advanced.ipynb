## ì‹¤ìŠµ - ìŠ¤íƒœí‹±(ì •ì ) ë©”ì„œë“œ

> ê³„ì‚°ê¸° classì¸ `Calculator`ë¥¼ ë§Œë“¤ì–´ë´…ì‹œë‹¤.

* ë‹¤ìŒê³¼ ê°™ì´ ì •ì  ë©”ì„œë“œë¥¼ êµ¬ì„±í•œë‹¤. 
* ëª¨ë“  ì •ì  ë©”ì„œë“œëŠ”, ë‘ ìˆ˜ë¥¼ ë°›ì•„ì„œ ê°ê°ì˜ ì—°ì‚°ì„ í•œ ê²°ê³¼ë¥¼ ë¦¬í„´í•œë‹¤.
* `a` ì—°ì‚°ì `b` ì˜ ìˆœì„œë¡œ ì—°ì‚°í•œë‹¤. (`a - b`, `a / b`)
    1. `add(a, b)` : ë§ì…ˆ

````python
# ì•„ë˜ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
class Calculator:
    
    @staticmethod
    def add(a,b):
        return a+b
    @staticmethod
    def sub(a,b):
        return a-b
    @staticmethod
    def mul(a,b):
        return a*b
    @staticmethod
    def div(a,b):
        return a/b
````

````python
# ì •ì  ë©”ì„œë“œë¥¼ í˜¸ì¶œí•´ë³´ì„¸ìš”.
a = Calculator()
print(a.add(2,4))
print(a.sub(2,4))
print(a.mul(2,4))
print(a.div(2,4))

#Calculator.add(2,4)
````


````python
# ì‚¬ëŒê³¼ ì‚¬ëŒì„ ê°™ì€ì§€ ë¹„êµí•˜ë©´, ì´ëŠ” ë‚˜ì´ê°€ ê°™ì€ì§€ ë¹„êµí•œ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ë„ë¡ ë§Œë“¤ì–´ë´…ì‹œë‹¤.
class Person:
    
    def __gt__(self,target):
        if self.age > target.age:
            return True
        else:
            return False
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __repr__(self):
        return(f'name: {self.name},age: {self.age}')

    def __lt__(self,target):
            if self.age < target.age:
                return True
            else:
                return False
````

````python
oldman = Person('Leon',50)
youngman = Person('Matilda', 15)
````

````python
oldman < youngman
# oldman.age > youngman.age
````
